### Get winners for league table and last 5 winners

Functions

```python
def decide_match_winner(home_score, away_score):
    """Function used to decide on the winner for a football match"""
    if home_score > away_score :
        return 'H'
    elif home_score == away_score:
        return 'D'
    else:
        return 'A'
```

Add winning column per match

```python
fixtures_df = seasons_dict['2022_2023']['fixtures']

fixtures_df = (
    fixtures_df
    .assign(
        home_score=fixtures_df.home_score.astype(float),
        away_score=fixtures_df.away_score.astype(float)
        )
    .assign(match_winner=fixtures_df.apply(lambda x: decide_match_winner(x.home_score, x.away_score), axis=1))
)

fixtures_df.head(3)
```
